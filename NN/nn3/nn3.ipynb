```python
%matplotlib inline
import importlib
import nn
importlib.reload(nn)
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from nn import NeuralNetwork
```

### Data loading

```python
data_step_train = pd.read_csv("dane_nn3/steps-large-training.csv")
data_step_train = data_step_train.drop(data_step_train.columns[0], axis=1)
data_step_train.head()
X_step_train = data_step_train['x'].values.reshape(-1, 1)
y_step_train = data_step_train['y'].values.reshape(-1, 1)

data_step_test = pd.read_csv("dane_nn3/steps-large-test.csv")
data_step_test = data_step_test.drop(data_step_test.columns[0], axis=1)
X_step_test = data_step_test['x'].values.reshape(-1, 1)
y_step_test = data_step_test['y'].values.reshape(-1, 1)

data_square_train = pd.read_csv("dane_nn3/square-large-training.csv")
data_square_train = data_square_train.drop(data_square_train.columns[0], axis=1)
data_square_train.head()
X_square_train = data_square_train['x'].values.reshape(-1, 1)
y_square_train = data_square_train['y'].values.reshape(-1, 1)

data_square_test = pd.read_csv("dane_nn3/square-large-test.csv")
data_square_test = data_square_test.drop(data_square_test.columns[0], axis=1)
data_square_test.head()
X_square_test = data_square_test['x'].values.reshape(-1, 1)
y_square_test = data_square_test['y'].values.reshape(-1, 1)

multimodal_train = pd.read_csv("dane_nn3/multimodal-large-training.csv")
multimodal_train.head()
X_multimodal_train = multimodal_train['x'].values.reshape(-1, 1)
y_multimodal_train = multimodal_train['y'].values.reshape(-1, 1)

multimodal_test = pd.read_csv("dane_nn3/multimodal-large-test.csv")
multimodal_test.head()
X_multimodal_test = multimodal_test['x'].values.reshape(-1, 1)
y_multimodal_test = multimodal_test['y'].values.reshape(-1, 1)
```

# Steps large dataset

### Mini-batch


```python
nn_step_batch = NeuralNetwork(X_step_train, y_step_train, [1, 10, 10, 1], activation_fun='sigmoid')
batch = 100
history_step_batch = nn_step_batch.train(learning_rate=0.01, epochs=int(1e3), validation_data=(X_step_test, y_step_test),
                                          mini_batch=True, batch_size=batch, momentum=0, stop_condition=1, report_interval=10)
history_step_batch = nn_step_batch.train(learning_rate=0.001, epochs=int(1e3), validation_data=(X_step_test, y_step_test),
                                          mini_batch=True, batch_size=batch, momentum=0, stop_condition=1, report_interval=10)
history_step_batch = nn_step_batch.train(learning_rate=0.0001, epochs=int(1e3), validation_data=(X_step_test, y_step_test),
                                          mini_batch=True, batch_size=batch, momentum=0, stop_condition=1, report_interval=10)
```

### Mini-batch with momentum

```python
nn_step_momentum = NeuralNetwork(X_step_train, y_step_train, [1, 10, 10, 1], activation_fun='sigmoid')
batch = 100
history_step_momentum = nn_step_momentum.train(learning_rate=0.01, epochs=int(1e3), validation_data=(X_step_test, y_step_test),
                                          mini_batch=True, batch_size=batch, momentum=0.7, stop_condition=1, report_interval=10)
history_step_momentum = nn_step_momentum.train(learning_rate=0.001, epochs=int(1e3), validation_data=(X_step_test, y_step_test),
                                          mini_batch=True, batch_size=batch, momentum=0.7, stop_condition=1, report_interval=10)
history_step_momentum = nn_step_momentum.train(learning_rate=0.0001, epochs=int(1e3), validation_data=(X_step_test, y_step_test),
                                          mini_batch=True, batch_size=batch, momentum=0.7, stop_condition=1, report_interval=10)
```

### Mini-batch with rmsprop

```python
nn_step_rmsprop = NeuralNetwork(X_step_train, y_step_train, [1, 10, 10, 1], activation_fun='sigmoid')
batch = 100
history_step_rmsprop = nn_step_rmsprop.train(learning_rate=0.01, epochs=int(3e3), validation_data=(X_step_test, y_step_test),
                                          mini_batch=True, batch_size=batch, optimization='rmsprop', stop_condition=1, report_interval=10)
```

```python
y_pred_batch = nn_step_batch.predict(X_step_test)
y_pred_momentum = nn_step_momentum.predict(X_step_test)
y_pred_rmsprop = nn_step_rmsprop.predict(X_step_test)

print("Mini-batch MSE: ", np.mean((y_step_test - y_pred_batch) ** 2))
print("Mini-batch with momentum MSE: ", np.mean((y_step_test - y_pred_momentum) ** 2))
print("Mini-batch with RMSprop MSE: ", np.mean((y_step_test - y_pred_rmsprop) ** 2))

x = np.linspace(min(X_step_test), max(X_step_test), 100).reshape(-1, 1)

plt.scatter(X_step_test, y_step_test, label="True")
plt.plot(x, nn_step_batch.predict(x), label="Mini-batch")
plt.plot(x, nn_step_momentum.predict(x), label="Mini-batch with momentum")
plt.plot(x, nn_step_rmsprop.predict(x), label="Mini-batch with RMSprop")
plt.legend()
plt.show()

# loss history plot
plt.figure(figsize=(10, 5))
plt.plot(history_step_batch, label='Mini-batch')
plt.plot(history_step_momentum, label='Mini-batch with momentum')
plt.plot(history_step_rmsprop, label='Mini-batch with RMSprop')
plt.yscale('log')
plt.legend()
plt.show()
```
